```python
# To plot pretty figures
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
#Histograms

data.hist(bins=50, figsize=(30, 50))
plt.show()
```

```python
data.plot(kind="scatter", x="energy", y="track_popularity",
    s=data["artist_popularity"], label="artist_popularity", figsize=(10,7), alpha=0.4, 
    c="artist_popularity", cmap=plt.get_cmap("jet"), colorbar=True,
)
plt.legend()
```

```python
data.plot(kind="scatter", x="danceability", y="track_popularity",
    s=data["artist_popularity"], label="artist_popularity", figsize=(10,7), alpha=0.4, 
    c="artist_popularity", cmap=plt.get_cmap("jet"), colorbar=True,
)
plt.legend()
```

```python
data2 = data.drop(["playlist_url", "track_id", "track_name", "album", "artist_id", "artist_name", "artist_genres"], axis=1)

corr_data = data2.corr()
corr_data
```

```python
import seaborn as sns
plt.figure(figsize=(12, 10))
sns.heatmap(corr_data, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.show()
```
