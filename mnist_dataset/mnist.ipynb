```python
#Loading data from mnist
import tensorflow as tf
import matplotlib.pyplot as plt
import seaborn as sns
import random
import pandas as pd
mnist = tf.keras.datasets.mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
print("x_train shape:", x_train.shape)
print("y_train shape:", y_train.shape)
print("x_test shape:", x_test.shape)
print("y_test shape:", y_test.shape)
```

```python
#Plotted the data
fig, axes = plt.subplots(2, 5, figsize=(12, 5))
for i, ax in enumerate(axes.flat):
    ax.imshow(x_train[i], cmap="gray")
    ax.set_title(f"Label: {y_train[i]}")
plt.show()
```

```python
#Plotted the Frequency of Digits in the training data set using Histogram:
plt.hist(y_train, bins=range(11), align="left", rwidth=0.8)
plt.xlabel("Digit")
plt.ylabel("Frequency")
plt.xticks(range(10))
plt.title("Distribution of Labels in Training Data")
plt.show()
```

```python
#statistics for pixel intensities (.describe for csv)
pixel_mean = x_train.mean()
pixel_std = x_train.std()
pixel_min = x_train.min()
pixel_max = x_train.max()

print(f"Mean Pixel Value: {pixel_mean}")
print(f"Standard Deviation of Pixel Values: {pixel_std}")
print(f"Minimum Pixel Value: {pixel_min}")
print(f"Maximum Pixel Value: {pixel_max}")
```

```python
# multiple variations of all digits (intensity for some digits vary, thicker/thinner)
for n in range(10):
    digit_to_display = n
    indices = [i for i, label in enumerate(y_train) if label == digit_to_display]

    plt.figure(figsize=(12, 5))
    for i, idx in enumerate(random.sample(indices, 5)):  # Display 5 random examples
        plt.subplot(1, 5, i + 1)
        plt.imshow(x_train[idx], cmap="gray")
        plt.title(f"Variation {i + 1}")
        plt.axis("off")
    plt.suptitle(f"Variations of Digit {digit_to_display}")
    plt.show()
```
